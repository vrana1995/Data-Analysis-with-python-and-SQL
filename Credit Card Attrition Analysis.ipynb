# Analysis of Attrition of Credit Card Customer.

### Here we have a dataset of Credit Card Customers of a Bank and task in hand is to analyse the data for possible reasons of Attrition/Churning of customers.
### We will use SQL for Analysing the data and draw conclusions from it.Also we will use Python to connect with SQL  to visualize our findings.

```python
import pandas as pd         ## importing pandas library.
import sqlalchemy as sq     ## importing sqlalchemy for connecting mysql and python.
```

```python
con = sq.create_engine('mysql+pymysql://root:user@localhost/my_database') ##engine will create a connection between SQL and PYTHON
```

```python
df = pd.read_csv('Credit_cards') ##Readig the dataset with Pandas
```

```python
df.head()
```

```python
df.to_sql(name = 'credit',con = con) ## Importing the dataset into SQL 
```

```python
## here we have established a connection with mysql and running a query to confirm it
sql= '''     
select *
from credit
limit 5'''
df.sql = pd.read_sql_query(sql,con)
df.sql
```

### The case study consists of 10127 customer Records  and 21 Features such as Age, Gender,Status etc.
### Our Objective -
### 1. Analyse the data and Predict which are the potential customers to get churned.
### 2. Categorise the data based on the findings.
### 3. Publish Findings so thta proactive measures can be taken to reduce Attrition.

```python
## Given that our data is clean and doesen't have null values, we can dive into exploring the data.
```

```python
## Distribution of ATTRITED customers based on age range.
sql= '''
SELECT case when Customer_Age< 20 then  '0-20'
            when Customer_Age between 20 and 30  then '20-30'
            when Customer_Age between 30 and 40  then  '30-40'
            when Customer_Age between 40 and 50  then  '40-50'
            when Customer_Age between 50 and 60  then  '50-60'
            when Customer_Age between 60 and 70  then  '60-70'
            when Customer_Age between 70 and 80  then  '70-80'
            when Customer_Age >80 then  'Above 80'
            end as Age_Range, count(*)
            from credit 
            where Attrition_Flag = 'Attrited Customer'
            group by Age_Range
            order by Age_Range;
'''
df.sql = pd.read_sql_query(sql,con)
df.sql
```

#### Finding - The number of customers attriting is highest in the age group of 40-50 followed by 50-60. Thus there is a possibility that people belonging to this range may have tendency to attrite.

```python
## Distribution of  Male vs female in existing and Attriting customers.
sql = '''select Gender ,count(*) Existing
   from credit
   where Attrition_Flag = 'Existing Customer'
   group by Gender;'''
df.sql = pd.read_sql_query(sql,con)
df.sql
```

```python
sql = '''
   select Gender ,count(*) Attrited
   from credit
   where Attrition_Flag = 'Attrited Customer'
   group by Gender;'''

df.sql = pd.read_sql_query(sql,con)
df.sql
```

#### Finding- Here we obeserve that numeber of female is higher in both attrited as well as existing customers.

```python
## Distribution based on Education Level of the attrited customers.
sql = '''
 select Education_Level,count(*) Attrited
   from credit 
   where Attrition_Flag = 'Attrited Customer'
   group by Education_Level
   order by Attrited DESC'''

df.sql = pd.read_sql_query(sql,con)
df.sql
```

#### Finding - here we see that the maximum number of attrited customers are from graduate followed by high school. we can infer that given the source of income is limited or close to none, they must be attriting due to lack of funds to compensate the credit card bill.

```python
## Distribution based on the Income 
sql = '''
select Income_Category, count(*) Attritted
from credit
where Attrition_Flag = 'Attrited Customer'
group by Income_Category
order by Income_Category,Attritted'''

df.sql = pd.read_sql_query(sql,con)
df.sql
```

#### Finding - we can see that for people having salary less than 40k ,attrition is the most, so may be income could be a reason for attrition.

```python
## Now based on the tenure of relationship with the bank
sql= '''
Select case when Months_on_book< 20 then  '0-20'
            when Months_on_book between 20 and 30  then '20-30'
            when Months_on_book between 30 and 40  then  '30-40'
            when Months_on_book between 40 and 50  then  '40-50'
            when Months_on_book between 50 and 60  then  '50-60'
            when Months_on_book> 60 then 'Above 60'
            end Months_with_Bank,count(*)
            from credit
            where  Attrition_Flag = 'Attrited Customer'
            group by  Months_with_Bank
            order  by Months_with_Bank'''

df.sql = pd.read_sql_query(sql,con)
df.sql
```

#### Finding -  We observe here that maximum attrition happen for people who have been with bank for 3 years or more.

```python
## Now  Based on the Martial status
sql = '''
 select Marital_Status, count(*) Attrited
            from credit
            where  Attrition_Flag = 'Attrited Customer'
            group by Marital_Status'''
df.sql = pd.read_sql_query(sql,con)
df.sql
```

#### Finding - Here we obeserve martial status is really not playing a role as attrition for Single is also comparable.

```python
## Based on the type of card Customer has
sql = '''
select Card_Category, count(*) Attrited
            from credit
            where  Attrition_Flag = 'Attrited Customer'
            group by Card_Category;'''

df.sql = pd.read_sql_query(sql,con)
df.sql
```

#### Finding - We observe that maximum attrition is happeing for the customers with Category Blue card. Also cards like Gold and platinum have negligble attrition. May be could be because of the benefits they are getting.

```python
## Based on Month Inactivity
sql = '''
select Months_Inactive_12_mon, count(*) Attrited
            from credit
            where  Attrition_Flag = 'Attrited Customer'
            group by Months_Inactive_12_mon
            order by Months_Inactive_12_mon;'''

df.sql = pd.read_sql_query(sql,con)
df.sql
```

#### Finding-  Based on this ,we can infer that people who are inactive for 2-3 months are most likely to attrite.

```python
## Now based on the credit card limit for customer
sql = '''
select case when Credit_Limit <5000 then '0-5000'
            when Credit_Limit between 5000 and 10000 then '5000-10000'
            when Credit_Limit between 10000 and 15000 then '10000-15000'
            when Credit_Limit between 15000 and 20000 then '15000-20000'
            when Credit_Limit between 20000 and 25000 then '20000-25000'
            when Credit_Limit between 25000 and 30000 then '25000-30000'
            when  Credit_Limit < 35000 then 'Above 35000'
            end as Credit_Range , count(*) Attrition
from credit

where Attrition_Flag = 'Attrited Customer'
group by Credit_Range
order by  Credit_Range DESC'''
df.sql = pd.read_sql_query(sql,con)
df.sql
```

#### Finding - We observe that for people with credit card limit less than 10000 are likely to attrite

### Findings -
### 1. The Possibility of customers attriting between the age of 40-50 years is highest followed by 50-60.
### 2. The Possibility of attrition of Female customers is higher than that of male customers.
### 3. We can infer that the maximum number of attrited customers are from graduates followed by high school.
### 4. The risk of attrition is higher for customers who have been with the bank for 3 years or more.
### 5. The Possibility for customers who are Blue Card Customer attriting is much more and also credit limit less than 10000..
