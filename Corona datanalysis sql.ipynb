```python
sql = '''
Select location,date,total_cases,total_deaths,(total_deaths/total_cases)*100 as DeathPercent
from covid_deaths
where location = 'india'
limit 50
 '''

df1.sql= pd.read_sql_query(sql,con) ## first case and likelihood of contracting.gettimg indioa specific but otherts also
df1.sql
```

```python
## looking total cases vs population 

sql='''Select location,date,total_cases,population,(total_cases/population)*100 as CasePercent
from covid_deaths
where location  = 'india'
'''

df1.sql= pd.read_sql_query(sql,con)
df1.sql
```

```python
## highest infection rate

sql='''Select location,population,max(total_cases)as highestInfection,max((total_cases/population))*100 as PercentPoulationinfected
from covid_deaths
group by location,population
order by Percentpoulationinfected desc
limit 51
'''

df1.sql= pd.read_sql_query(sql,con)
df1.sql
```

```python
# highest deaths by country,continent is not null
sql='''Select location,max(total_deaths) as highestDeath 
from covid_deaths
where continent is not null
group by location,population
order by highestDeath  desc
limit 20
'''

df1.sql= pd.read_sql_query(sql,con)
df1.sql
```

```python
## highest death   by continent
sql='''Select continent,max(total_deaths) as Deathcount 
from covid_deaths
where continent is not null
group by continent
order by Deathcount desc

'''

df1.sql= pd.read_sql_query(sql,con)
df1.sql
```

```python
# new cases and deaths everyday globally 
sql='''
select date, sum(new_cases) as Daily_cases, sum(total_deaths) as Daily_deaths,
from covid_deaths
where continent is not null
group by date
order by 2
limit 50
'''
df1.sql= pd.read_sql_query(sql,con)
df1.sql
```

```python
# % death all over world
sql='''
select  sum(total_cases) as Daily_cases, sum(total_deaths) as Daily_deaths,  (sum(total_deaths)/sum(total_cases))*100
as WorldDeathPercent
from covid_deaths
where continent is not null

limit 50
'''
df1.sql= pd.read_sql_query(sql,con)
df1.sql
```

```python
## looking at total population vs vaccinated we will join two tables
sql= '''
select *
from covid_deaths cd 
join covid_vac cv
on cd.location = cv.location 
and cd.date= cv.date'''

df1.sql= pd.read_sql_query(sql,con)
df1.sql.head()
```

```python
sql = '''
select cd.continent,cd.location,cd.date,cd.population,cv.new_vaccinations
from covid_deaths cd 
join covid_vac cv
on cd.location = cv.location 
and cd.date= cv.date
where cd.continent is not null
order  by 2
limit 50'''

df1.sql= pd.read_sql_query(sql,con)
df1.sql.head()
```

```python
## rolling count of vaccination
sql = '''
select cd.continent,cd.location,cd.date,cd.population,cv.new_vaccinations,
sum(cv.new_vaccinations) over(partition by cd.location order by cd.location,cd.date ) as Rolling_count_vacc
from covid_deaths cd 
join covid_vac cv
on cd.location = cv.location 
and cd.date= cv.date
where cd.continent is not null and cd.location = 'india'
order  by 2 
'''

df1.sql= pd.read_sql_query(sql,con)
df1.sql.tail(20)
```

```python
## %  population vs vaccination per date for india

sql= '''
with cte as(select cd.continent,cd.location,cd.date,cd.population,cv.new_vaccinations,
sum(cv.new_vaccinations) over(partition by cd.location order by cd.location,cd.date ) as Rolling_count_vacc
from covid_deaths cd 
join covid_vac cv
on cd.location = cv.location 
and cd.date= cv.date
where cd.continent is not null 
order  by 2) 

select continent,location,date,population, (Rolling_count_vacc/population)*100 as People_vaccinated_perct
from cte
where location ='india'
'''
df1.sql= pd.read_sql_query(sql,con)
df1.sql.tail(50)
```
