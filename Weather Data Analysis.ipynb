```python
## Now using the graphs to analyse the data.
## 1. we compare Temperature in each month and wind speed.
```

```python
plt.rcParams["figure.figsize"] = [10,5]
Data.plot(x='Month',y='Temp_C'),
Data.plot(x='Month',y='Wind Speed_km/h');
```

We find that for January month Wind speed is  the highest and also temperature is the lowest reaching Beyond -20C. whereas for the months of june to september, temperature is high due to less speed of Wind.
Here we can infer that Temperature and Wind are inversely related to each other.
